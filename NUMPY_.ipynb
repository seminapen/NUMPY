```python
import numpy as np  # NumPy kütüphanesini np olarak import ediyoruz

# Bir NumPy dizisi oluşturuyoruz ve yazdırıyoruz
print(np.array([2, 4, 6, 8, 10, 12]))

# Başka bir NumPy dizisi oluşturuyoruz
k = np.array([1, 2, 3, 4, 5, 6])
# Dizinin tipini kontrol ediyoruz
type(k)
# Diziyi yazdırıyoruz
print("\n", k)

# 5 elemanlı, elemanları sıfır olan bir NumPy dizisi oluşturuyoruz
np.zeros(5, dtype=np.float32)
# Yukarıdaki diziyi değişkene atayıp yazdırıyoruz
b = np.zeros(5, dtype=np.float32)
print("\n", b)

# 4 elemanlı, elemanları bir olan bir NumPy dizisi oluşturuyoruz ve yazdırıyoruz
print("\n", np.ones(4, dtype=int))

# 2x4 boyutlarında, elemanları bir olan bir NumPy dizisi oluşturuyoruz ve yazdırıyoruz
print("\n", np.ones((2, 4), dtype=int))

# 2x4 boyutlarında, tüm elemanları 6 olan bir NumPy dizisi oluşturuyoruz ve yazdırıyoruz
c = np.full((2, 4), 6)
print("\n", c)

# 22'den başlayarak 66'ya kadar (66 hariç) 4 artarak giden bir NumPy dizisi oluşturuyoruz ve yazdırıyoruz
print(np.arange(22, 66, 4))

# 0 ile 2 arasında (2 dahil) 20 eşit parça oluşturarak bir NumPy dizisi oluşturuyoruz ve yazdırıyoruz
print("\n",np.linspace(0, 2, 20))

# 2 ile 12 arasında (12 hariç) rastgele tam sayılar içeren 3x3 boyutlarında bir NumPy dizisi oluşturuyoruz ve yazdırıyoruz
print("\n",np.random.randint(2, 12, (3, 3)))

# Ortalaması 14 ve standart sapması 5 olan normal dağılımdan rastgele sayılar içeren 4x4 boyutlarında bir NumPy dizisi oluşturuyoruz ve yazdırıyoruz
print("\n",np.random.normal(14, 5, (4, 4)))



```

```python
# Bir NumPy dizisi oluşturuyoruz
a = np.array([7, 8, 9, 10, 11, 12, 13])

# Dizinin kaç boyutlu olduğunu yazdırıyoruz
print(a.ndim)  # ndim, dizinin boyut sayısını döndürür. Bu dizi tek boyutludur (vektördür).

# Dizinin şeklini (boyutlarını) yazdırıyoruz
print("\n", a.shape)  # shape, dizinin her boyuttaki eleman sayısını döndürür. Bu dizi 1 boyutlu ve 7 elemanlıdır.

# Dizinin toplam eleman sayısını yazdırıyoruz
print("\n", a.size)  # size, dizideki toplam eleman sayısını döndürür. Bu dizide 7 eleman vardır.
```

```python
# 1. Tek boyutlu diziler oluşturuluyor
m = np.array([4, 5, 6, 7])
n = np.array([1, 2, 3, 5])

# 2. axis parametresi verilmezse, varsayılan olarak axis=0 kullanılır
result = np.concatenate([m, n])
print(result)
# Açıklama: Bu satırda, m ve n tek boyutlu dizileri concatenate fonksiyonu kullanılarak birleştiriliyor.
# Sonuç, [4 5 6 7 1 2 3 5] şeklinde olur ve diziler birbirinin ardına eklenmiş olur.

# 3. İki boyutlu diziler oluşturuluyor
m = np.array([[4, 5, 6, 7]])
n = np.array([[1, 2, 3, 5]])

# 4. axis=0 parametresi ile iki boyutlu diziler birleştiriliyor
result = np.concatenate([m, n], axis=0)
print("\n", result)
# Açıklama: Bu satırda, m ve n iki boyutlu dizileri (her biri tek bir satır içeriyor) concatenate fonksiyonu ile axis=0
# (yani dikey eksen) üzerinde birleştiriliyor. Sonuç, alt alta eklenmiş iki boyutlu bir dizi olur:
# [[4 5 6 7]
#  [1 2 3 5]]

# 5. İki boyutlu diziler oluşturuluyor (tekrar)
m = np.array([[4, 5, 6, 7]])
n = np.array([[1, 2, 3, 5]])

# 6. axis=1 parametresi ile iki boyutlu diziler birleştiriliyor
result = np.concatenate([m, n], axis=1)
print("\n", result)
# Açıklama: Bu satırda, m ve n iki boyutlu dizileri concatenate fonksiyonu ile axis=1 (yani yatay eksen) üzerinde
# birleştiriliyor. Sonuç, yan yana eklenmiş iki boyutlu bir dizi olur:
# [[4 5 6 7 1 2 3 5]]

```
